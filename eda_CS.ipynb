```python
# Group by 'Organization_Name' and calculate mean feedback, completion rates, and assignment completion for each
workshop1_grouped = new_data_cleaned.groupby('Organization_Name').agg({
    'Feedback_Day1': 'mean',
    'Feedback_Day2': 'mean',
    'Feedback_Day3': 'mean',
    'Feedback_Day4': 'mean',
    'Feedback_Day5': 'mean',
    'Feedback_Day6': 'mean',
    'Feedback_Day7': 'mean',
    'Feedback_Day8': 'mean',
    'Feedback_Day9': 'mean',
    'Assignment_Completion_Percentage': 'mean',
    'Completion_1st_2nd_Session': 'mean',
    'Completion_3rd_Session_25Jan2024': 'mean',
    'Completion_4th_Session_5Feb2024': 'mean',
    'Completion_5th_Session_22Feb2024': 'mean',
    'Completion_6th_Session_29Feb2024': 'mean',
    'Completion_7th_Session_12Mar2024': 'mean',
    'Completion_8th_Session_19Mar2024': 'mean',
    'Completion_9th_Session_22Mar2024': 'mean',
    'Completion_9th_Session_22Apr2024': 'mean'
})

# Calculate the overall average for feedback and completion rates for each hospital
workshop1_grouped['Avg_Feedback_Score'] = workshop1_grouped[['Feedback_Day1', 'Feedback_Day2', 'Feedback_Day3', 'Feedback_Day4', 'Feedback_Day5', 'Feedback_Day6', 'Feedback_Day7', 'Feedback_Day8', 'Feedback_Day9']].mean(axis=1)
workshop1_grouped['Avg_Completion_Rate'] = workshop1_grouped[['Completion_1st_2nd_Session', 'Completion_3rd_Session_25Jan2024', 'Completion_4th_Session_5Feb2024', 'Completion_5th_Session_22Feb2024', 'Completion_6th_Session_29Feb2024', 'Completion_7th_Session_12Mar2024', 'Completion_8th_Session_19Mar2024', 'Completion_9th_Session_22Mar2024', 'Completion_9th_Session_22Apr2024']].mean(axis=1)

# Add the overall assignment completion percentage
workshop1_grouped['Avg_Assignment_Completion'] = workshop1_grouped['Assignment_Completion_Percentage']

# Combine metrics into a single performance score (e.g., weighted sum or average)
workshop1_grouped['Performance_Score'] = workshop1_grouped[['Avg_Feedback_Score', 'Avg_Completion_Rate', 'Avg_Assignment_Completion']].mean(axis=1)

# Sort by performance score to find the top-performing hospital
top_hospital_workshop1 = workshop1_grouped.sort_values('Performance_Score', ascending=False).head(1)

top_hospital_workshop1
```
